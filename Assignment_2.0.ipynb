```python
X=df[['carlength','carwidth','carheight','enginesize','stroke','horsepower']]
Y=df['price']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
len(X_test)
```

```python
clf=LinearRegression()
```

```python
clf.fit(X_train , Y_train)
```

```python
clf.predict(X_test)
```

```python
Y_test
```

```python
clf.score(X_test, Y_test)
```

```python
pred_price=clf.intercept_+clf.coef_[0]*190.9+clf.coef_[1]*70.3+clf.coef_[2]*54.9+clf.coef_[3]*183+clf.coef_[4]*3.64+clf.coef_[5]*123
print(pred_price)
```
